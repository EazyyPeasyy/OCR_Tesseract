# Importações

```python
import pytesseract
import numpy as np
import cv2 #OpenCV
from google.colab.patches import cv2_imshow
```

# **LENDO IMAGEM E MOSTRANDO ELA**

```python
img = cv2.imread('/content/teste.png')
cv2_imshow(img)
```

# **TRANSFORMANDO IMAGEM BGR PARA RGB**

```python
rgb = cv2.cvtColor(img, cv2.COLOR_BGR2RGB)
cv2_imshow(rgb)
```

# **APLICANDO TESSERACT PARA LER A IMAGEM E TRANSFORMAR EM STRING**

```python
texto = pytesseract.image_to_string(rgb)
print(texto)
```

# **FAZENDO COM QUE SUPORTE IDIOMA PORTUGUÊS**

```python
texto = pytesseract.image_to_string(img, lang="por")
print(texto)
```

# **RECEBNDO UMA IMAGEM -> TRANSFORMANDO EM RGB -> E PRINTANDO**

```python
img = cv2.imread("Captura de tela 2022-03-31 035215.png")
img_rgb = cv2.cvtColor(img, cv2.COLOR_BGR2RGB)
cv2_imshow(img_rgb)
```

# **CRIANDO UMA CONFIG_TESSERACT PARA PASSARMOS VÁRIOS PARÂMETROS E LOGO EM SEGUIDA CONVERTENDO DE IMAGEM PARA STRING**

```python
config_tesseract = '--tessdata-dir tessdata --psn 6'
text = pytesseract.image_to_string(img_rgb, lang="por", config = config_tesseract)
print(text)
```

# **REPETINDO O TEXTO MAS COM UMA IMAGEM DIFERENTE**

```python
img = cv2.imread("Saída.png")
img_rgb = cv2.cvtColor(img, cv2.COLOR_BGR2RGB)
cv2_imshow(img_rgb)
```

```python
config_tesseract = '--tessdata-dir tessdata --psn 7'
text = pytesseract.image_to_string(img_rgb, lang="por", config = config_tesseract)
print(text)
```

# Seleção de textos

```python
from pytesseract import Output
```

```python
img = cv2.imread('/content/name.png')
img_rgb = cv2.cvtColor(img, cv2.COLOR_BGR2RGB )
cv2_imshow(img_rgb)
```

```python
config_tesseract = "--tesseract-dir tessdata"
result = pytesseract.image_to_data(img_rgb, config=config_tesseract, lang="por", output_type=Output.DICT)
result
```
